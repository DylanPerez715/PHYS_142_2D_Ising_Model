```python
STEPS = 200_000
N = 100
J = 1
KB = 1
T = 2
#lattice_spins = np.ones((N, N))
lattice_spin_random = 2 * (np.random.randint(2, size=(N, N)) - 0.5)


        
plt.figure()
plt.imshow(lattice_spin_random, cmap="RdYlBu")
# show gridlines
for i in range(N):
    plt.axhline(i + 0.5, color="black", lw=0.1)
    plt.axvline(i + 0.5, color="black", lw=0.1)
plt.show()
```

```python
num_accept=0

def Hastings (B, T ,lattice, steps,shape):
    
    m_values=[]
    num_accept=0
    for t in tqdm.tqdm(range(steps)):
        i, j = np.random.randint(shape), np.random.randint(shape)
   
        delta_energy = 0
        for k, l in [(-1, 0), (1, 0), (0, 1), (0, -1)]:
            i_neigh = i + k if i + k < shape else 0
            j_neigh = j + l if j + l < shape else 0
            delta_energy += -J * -2 * lattice[i, j] * lattice[i_neigh, j_neigh] - B * lattice[i,j]
        if delta_energy <= 0:
            lattice[i, j] *= -1
            num_accept += 1
        elif delta_energy > 0:
            prob = np.exp(-delta_energy / (T))
            if np.random.random() < prob:
                lattice[i, j] *= -1
                num_accept += 1
        m_values.append(np.mean(lattice))
        
    return lattice, m_values

```

```python
temp_1, mean_1 =Hastings(-1,1,lattice_spin_random, 200_000, 100)


plt.figure()
plt.imshow(temp_1, cmap="RdYlBu")

for i in range(N):
    plt.axhline(i + 0.5, color="black", lw=0.1)
    plt.axvline(i + 0.5, color="black", lw=0.1)
plt.show()
```
